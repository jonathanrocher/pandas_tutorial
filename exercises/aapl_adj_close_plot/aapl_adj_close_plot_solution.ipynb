Question 1
----------
The loading of all the historical data is done below. It provides a `DataFrame` with the columns `Open`, `Close`, `High`, `Low`, `Volume` and `Adj Close` and has the `Date` column as its index.  Extract the column for adjusted close into a timeseries, called `adj_close`.

```python
aapl_df = pd.read_table("HistoricalAPPL_1980Dec12_2014Jul8.csv", sep=",", index_col=0)
```

```python
aapl_df.head()
```

```python
adj_close = aapl_df["Adj Close"]
```

```python
adj_close.head()
```

Question 2
----------
Try plotting that timeseries directly. You will see that the time is increasing going toward the left which is unnatural. Reverse the timeseries to display values from the old dates to the new dates. This can be done using one of the Series methods or just with indexing.

```python
adj_close.plot()
```

```python
# The fastest way to sort the series by date is to rely on the fact 
# that the order is correct and just reverse the Series
rev_adj_close = adj_close[::-1]
# Another (slower) solution is to sort the Series based on its index:
rev_adj_close = adj_close.sort_index()
```

```python
rev_adj_close.head()
```

```python
aapl_df = pd.read_table("HistoricalAPPL_1980Dec12_2014Jul8.csv", sep=",", index_col=0, parse_dates=True)
```

```python
adj_close = aapl_df["Adj Close"]
print(adj_close.index.dtype)
```

```python
rev_adj_close = adj_close[::-1]
```

Question 4
----------
Plot the timeseries to generate a plot with a title, a label for the y axis and no label for the x axis since the values are obvious. Save the figure to a png file. Make sure that the png generated is large enough for reading values precisely. 

```python
rev_adj_close.plot(style="g", figsize=(12, 8))
xlabel("")
ylabel("Adj. Close")
title("AAPL adjusted close stock price over its history")
savefig("aapl_adj_close.png")
```
